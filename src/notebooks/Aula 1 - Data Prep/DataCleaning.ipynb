## Data Cleaning e Data Preparation

### Importando as Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Criando um Dataset

```python
dados = pd.read_csv(r'https://raw.githubusercontent.com/ribeiromatheus/imdb-dataset/master/movies.csv')
```

```python
dados
```

# Tratamento de Nulos

```python
dados.describe()
```

```python
dados.isnull().sum()
```

Existem algumas estratégias que podemos usar

Para variáveis categóricas: Podemos usar a moda

Para variáveis contínuas: Podemos usar a média ou a mediana

Lembrando que existem casos onde o melhor será remover a linha inteira.

```python
dados['director_facebook_likes'] = dados['director_facebook_likes'].fillna(dados['director_facebook_likes'].mean())
```

```python
dados
```

```python
dados = dados.dropna()
```

```python
dados
```

# Tratamento de Duplicatas

```python
dados_duplicados = dados[['color', 'language']]
```

```python
dados_duplicados
```

```python
dados_duplicados.drop_duplicates().count()
```

```python
dados_duplicados.drop_duplicates()
```

# Tratamento de String

```python
dados.loc[:, 'director_name'] = dados['director_name'].str.replace(' ', '_').str.lower()
```

```python
dados
```
