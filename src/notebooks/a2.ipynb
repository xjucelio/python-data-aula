Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
base = pd.read_csv('../../data/train.csv')
```

```python
base.head(5)
```

Vamos entender qual a relacao da idade com o fator de sobrevivencia no titanic

```python
# Grafico basico - linhas

plt.plot(base.Age)
```

```python
# grafico de barras

plt.bar(base.Survived,base.Age.mean())
```

```python
# barras horizontais

plt.barh(base.Survived, base.Age)
```

```python
# grafico de dispensao

plt.scatter(base.Survived, base.Age)
```

```python
# removendo valores nulos

base2 = base.dropna(axis=0, subset=['Age'])
```

```python
plt.hist(base2.Age)
```

montando a tela do grafico

```python
plt.figure(figsize=(20, 10))
```

editando a tela

```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
```

Plotando o grafico

```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2.Age)
```

mudando a cor do grafico

```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2.Age, color='green')
```

separando os graficos (um para sobreviventes e outro nao sobreviventes)

```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2[base2['Survived']==0].Age, color='red')
```

```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2[base2.Survived==1].Age, color='blue')
```

visualizando as duas opcoes na mesma tela

```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2[base2.Survived==0].Age, color='red')
plt.hist(base2[base2.Survived==1].Age, color='blue')
```

```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2[base2.Survived==0].Age, color='red', alpha= 0.5)
plt.hist(base2[base2.Survived==1].Age, color='blue', alpha= 0.5)
```

```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2[base2.Survived==0].Age, color='red', alpha= 0.5, bins=30)
plt.hist(base2[base2.Survived==1].Age, color='blue', alpha= 0.5, bins=30)
```

```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2[base2.Survived==0].Age, color='red', alpha= 0.5, bins=30, edgecolor='black')
plt.hist(base2[base2.Survived==1].Age, color='blue', alpha= 0.5, bins=30, edgecolor='black')
```

```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2[base2.Survived==0].Age, color='red', label='Nao sobreviveu', alpha= 0.5, bins=30, edgecolor='black')
plt.hist(base2[base2.Survived==1].Age, color='blue', label='Sobreviveu', alpha= 0.5, bins=30, edgecolor='black')
plt.legend(loc='upper right')
```

```python
plt.figure(figsize=(20, 10))

plt.title('Histograma de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.hist(base2[base2.Survived==0].Age, color='red', label='Nao sobreviveu', alpha= 0.5, bins=30, edgecolor='black')
plt.hist(base2[base2.Survived==1].Age, color='blue', label='Sobreviveu', alpha= 0.5, bins=30, edgecolor='black')
plt.legend(loc='upper right')
plt.savefig('../../data/graficoidade.png')
```
