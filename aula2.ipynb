```python
def funcaopedro(texto = 'Pedro Alves'): # texto padrao se nao passado arg
    """ Retorna a primeira e a ultima letra do texto passado + a string - teste """
    return(texto[0]+texto[-1]+' - Final')
```

```python
funcaopedro()
```

Funcao lambda

```python
funcaopedro = lambda x: x[0]+x[-1]+' - Final'
```

```python
for i in base.index:
    if base['Name'][i][0] == 'B':
        base.loc[i, 'ColunaNova3'] = funcaopedro(base['Name'][i])
    elif base['Name'][i][0] == 'C':
        base.loc[i, 'ColunaNova3'] = funcaopedro(base['Name'][i])
    else:
        pass
```

```python
base
```

```python
# usamos o operador Try quando rodar um codigo com tolerancia a erros
# definindo novamente a pedro

def funcaopedro(texto):
    return (texto[0]+texto[-1]+ ' - Final')

# fazendo a operacao repetido para a coluna inteira

for i in base.index:
    try:
        base.loc[i, 'ColunaNova5'] = funcaopedro(base['Cabin'][i])
    except:
        base.loc[i, 'ColunaNova5']  = funcaopedro('Pedro Alves')
```
