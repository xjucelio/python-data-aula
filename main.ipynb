```python
import numpy as np

idades = [30,5,10,7]
idades

```

```python
# min = minimo
np.min(idades)
```

```python
# max = maximo
np.max(idades)
```

```python
# mean = media
np.mean(idades)
```

```python
# std = desvio padrao
np.std(idades)
```

```python
# sum = soma
np.sum(idades)
```

```python
# isnan - retorna true quando o e vazio
# nan - retorna um valor vazio
idades[1]
```

```python
np.nan
```

```python
idades.append(np.nan) # type: ignore
idades
```

```python
np.isnan(idades[4])
```

```python
idades.remove(np.nan) # type: ignore
```

```python
idades
```

```python
# quantile - retorna quantil da sua serie de dados
np.quantile(idades, 0.75)
```
